```python
from alpha_vantage.timeseries import TimeSeries
from pprint import pprint
ts = TimeSeries(key='CONXKCDKUI83AHB5', output_format='pandas')
df, meta_df = ts.get_intraday(symbol='MSFT',interval='1min', outputsize='full')
pprint(df.head(2))
```
